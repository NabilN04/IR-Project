```python
import pandas as pd

# Load
df = pd.read_csv("/content/BMW Dataset.csv")



# Convert Year into integer as its a numeric period
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")




# Missing values handling
df["Mileage_KM"] = df["Mileage_KM"].fillna("Unknown")
df = df.dropna(subset=["Model"])



# Remove duplicates
df = df.drop_duplicates()



# Fix numeric columns
numeric_cols = ["Engine_Size_L", "Mileage_KM", "Price_USD", "Sales_Volume"]


df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")



# Remove negative or impossible values
df = df[df["Price_USD"] >= 0]
df = df[df["Mileage_KM"] >= 0]
df = df[df["Sales_Volume"] >= 0]







# Final check
print(df.info())
print(df.head())
```

```python
# Save cleaned file
df.to_csv("Cleaned_BMW_Dataset.csv", index=False)
```
